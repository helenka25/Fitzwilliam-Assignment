```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Top 100 Crypto Currencies 2022 Translated.csv")
```

```python
crypto_99_2022=crypto_100_2022.dropna()
crypto_99_2022
```

```python
crypto_99_2022=crypto_99_2022.replace({'\ N/A':''}, regex = True)
crypto_99_2022
```

```python
crypto_99_2022=crypto_99_2022.replace({'\$':''}, regex = True)
crypto_99_2022
```

```python
crypto_99_2022=crypto_99_2022.replace({'\%':''}, regex = True)
crypto_99_2022
```

```python
crypto_99_2022.info()
```

```python
crypto_99_2022 = crypto_99_2022.replace(" ", "")
crypto_99_2022
```

```python
crypto_99_2022['Price'] = crypto_99_2022['Price'].astype(str).str.replace(',', '')
crypto_99_2022
```

```python
crypto_99_2022["Price"] = pd.to_numeric(crypto_99_2022["Price"])
crypto_99_2022
```

```python
crypto_99_2022["Changes 24H"] = pd.to_numeric(crypto_99_2022["Changes 24H"])
crypto_99_2022
```

```python
crypto_99_2022["Changes 7D"] = pd.to_numeric(crypto_99_2022["Changes 7D"])
crypto_99_2022
```

```python
crypto_99_2022["Changes 30D"] = pd.to_numeric(crypto_99_2022["Changes 30D"])
crypto_99_2022
```

```python
crypto_99_2022=crypto_99_2022.replace({'\ B':''}, regex = True)
crypto_99_2022
```

```python
crypto_99_2022["Changes 1Y"] = pd.to_numeric(crypto_99_2022["Changes 1Y"])
crypto_99_2022
```

```python
crypto_99_2022["Available Supply"] = pd.to_numeric(crypto_99_2022["Available Supply"])
crypto_99_2022
```

```python
crypto_99_2022.info()
```

```python
crypto_99_2022.isna().sum()
```

```python
rank_name = crypto_99_2022.iloc[:, [0,1]]
rank_name
```

```python
crypto_99_2022_numbers = crypto_99_2022.loc[:, 'Price':'Available Supply']
crypto_99_2022_numbers
```

```python
from sklearn import preprocessing

pt = preprocessing.PowerTransformer(method='yeo-johnson')
pt.fit(crypto_99_2022_numbers)
yeo_jackson = pt.transform(crypto_99_2022_numbers)
yeo_jackson=pd.DataFrame(yeo_jackson, columns=['Price', 'Changes 24H', 'Changes 7D', 'Changes 30D', 'Changes 1Y', 'Market Cap', 'Volume 24H', 'Available Supply'])
yeo_jackson
```

```python
yeo_jackson.isna().sum()
```

```python
yeo_jackson= pd.concat([rank_name, yeo_jackson], axis = 1)
yeo_jackson
```

```python
yeo_jackson_new.rename(columns={'Changes 24H': 'Changes_24H', 'Changes 7D': 'Changes_7D', 'Changes 30D': 'Changes_30D', 'Changes 1Y': 'Changes_1Y', 'Market Cap': 'Market_Cap', 'Volume 24H': 'Volume_24H', 'Available Supply': 'Available_Supply'}, inplace=True)
yeo_jackson_new
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
model = ols(formula = 'Ranking ~ Price + Changes_24H + Changes_7D + Changes_30D + Changes_1Y + Market_Cap + Volume_24H + Available_Supply', data=yeo_jackson_new).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
model = ols(formula = 'Ranking ~ Price', data=yeo_jackson_new).fit()
print(model.summary())
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
model = ols(formula = 'Ranking ~ Price + Available_Supply', data=yeo_jackson_new).fit()
print(model.summary())
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
model = ols(formula = 'Ranking ~ Price + Available_Supply + Market_Cap', data=yeo_jackson_new).fit()
print(model.summary())
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
model = ols(formula = 'Ranking ~ Price + Available_Supply + Market_Cap + Changes_1Y', data=yeo_jackson_new).fit()
print(model.summary())
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
model = ols(formula = 'Ranking ~ Price + Available_Supply + Market_Cap + Changes_1Y + Volume_24H', data=yeo_jackson_new).fit()
print(model.summary())
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```
