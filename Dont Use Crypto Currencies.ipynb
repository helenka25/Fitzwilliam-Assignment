```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crypto_99_2022=crypto_100_2022.dropna()
crypto_99_2022
```

```python
crypto_99_2022=crypto_99_2022.replace({'\ N/A':''}, regex = True)
crypto_99_2022
```

```python
crypto_99_2022=crypto_99_2022.replace({'\$':''}, regex = True)
crypto_99_2022
```

```python
crypto_99_2022=crypto_99_2022.replace({'\%':''}, regex = True)
crypto_99_2022
```

```python
crypto_99_2022.info()
```

```python
crypto_99_2022 = crypto_99_2022.replace(" ", "")
crypto_99_2022
```

```python
crypto_99_2022['Price'] = crypto_99_2022['Price'].astype(str).str.replace(',', '')
crypto_99_2022
```

```python
crypto_99_2022["Price"] = pd.to_numeric(crypto_99_2022["Price"])
crypto_99_2022
```

```python
crypto_99_2022["Changes 24H"] = pd.to_numeric(crypto_99_2022["Changes 24H"])
crypto_99_2022
```

```python
crypto_99_2022["Changes 7D"] = pd.to_numeric(crypto_99_2022["Changes 7D"])
crypto_99_2022
```

```python
crypto_99_2022["Changes 30D"] = pd.to_numeric(crypto_99_2022["Changes 30D"])
crypto_99_2022
```

```python
crypto_99_2022=crypto_99_2022.replace({'\ B':''}, regex = True)
crypto_99_2022
```

```python
crypto_99_2022["Changes 1Y"] = pd.to_numeric(crypto_99_2022["Changes 1Y"])
crypto_99_2022
```

```python
crypto_99_2022["Available Supply"] = pd.to_numeric(crypto_99_2022["Available Supply"])
crypto_99_2022
```

```python
rank_name = crypto_99_2022.iloc[:, [0,1]]
rank_name
```

```python
crypto_99_2022_numbers = crypto_99_2022.loc[:, 'Price':'Available Supply']
crypto_99_2022_numbers
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(crypto_99_2022_numbers)
scaler.scale = np.std(crypto_99_2022_numbers, axis=0, ddof=1)
crypto_99_2022_stnd = pd.DataFrame(data = scaler.transform(crypto_99_2022_numbers), columns = crypto_99_2022_numbers.columns )
crypto_99_2022_stnd.head()
```

```python
crypto_99_2022_stnd.isna().sum()
```

```python
full_crypto_scld_new.rename(columns={'Changes 24H': 'Changes_24H', 'Changes 7D': 'Changes_7D', 'Changes 30D': 'Changes_30D', 'Changes 1Y': 'Changes_1Y', 'Market Cap': 'Market_Cap', 'Volume 24H': 'Volume_24H', 'Available Supply': 'Available_Supply'}, inplace=True)
full_crypto_scld_new
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
model = ols(formula = 'Price ~ Changes_24H + Changes_7D + Changes_30D + Changes_1Y + Market_Cap + Volume_24H + Available_Supply', data=full_crypto_scld_new).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

```python
model = ols(formula = 'Price ~ Market_Cap', data=full_crypto_scld_new).fit()
print(model.summary())
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```
